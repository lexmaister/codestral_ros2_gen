```python
from pathlib import Path
import pandas as pd
import numpy as np
from codestral_ros2_gen.utils.metrics_handler import MetricsHandler
from codestral_ros2_gen import get_project_root


mh = MetricsHandler(metrics_file=get_project_root() / Path('tests/data/sample_metrics.jsonl'))
mh.load_metrics()
mh.metrics_df
```

```python
rep_1 = mh.metrics_df.groupby('iteration')[['success', 'attempt', 'tests_passed']].apply('max')
rep_1['success'] = rep_1['success'].map({True: '', False: ''})
rep_1.columns = ('success', 'attempts', 'tests_passed')
rep_1
```

```python
rep2 = mh.metrics_df.groupby('iteration')[['attempt_time', 'total_tokens']].apply('mean').astype(int)
rep2.columns = ('avg attempt time, s', 'avg total tokens')
rep2
```

```python
report = pd.concat([rep_1, rep2], axis=1)
print(report.to_markdown(tablefmt='pretty'))
```
